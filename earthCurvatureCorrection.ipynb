```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import math as m
%matplotlib qt
```

```python
im = Image.open('N037E022/ALPSMLC30_N037E022_DSM.tif')
DEM = np.array(im)
plt.figure(figsize=(10,10));plt.imshow(DEM); #im.show()
print('Size:',DEM.shape,'\nMax:',DEM.max(),' Min:',DEM.min())
```

```python
Rearth = 6371e3 #m
Dearth = m.pi * 2 * Rearth

#For a point distance X and height h in the DEM
h = 1000 #m
hdiff_all = []
for X in range(100,2001,10):

    theta = 2*m.pi * X / Dearth  
    h1 = (Rearth/m.cos(theta))-Rearth
    h2 = h - h1
    hnew = h2*m.cos(theta)
    hdiff_all.append(hnew-h)
    Xnew = m.tan(theta)*(Rearth + hnew)

plt.plot(np.arange(100,2001,10),hdiff_all)
```
